```python
def plot_one_ts(df,col,index, color = 'black'):
    sns.lineplot(data=df, 
        x="Timestamp", 
        y=col, 
        color=color
    )
    sns.despine()
    plt.show()
    
def plot_multiseries_ts(df, cols, figsz=(18, 10)):
    sz = len(cols) 
    if sz % 2 ==0: 
        num_rows = sz //2 
        fig, axes = plt.subplots(num_rows, 2, figsize=figsz)
    else:
        num_rows = sz // 2 + 1
        fig, axes = plt.subplots(num_rows, 2, figsize=figsz) 

    if sz ==2: ax=axes[:,None]
    else: ax=axes
    
    step_sub_plot = 0 
    var =0 
    for col in cols: 
        color = sample(list(mcolors.TABLEAU_COLORS.keys()), k=1)
        sns.lineplot(x="Timestamp", 
                     y=col,
                     data= df.iloc[-1000:],
                     ax=ax[var, step_sub_plot], 
                     color=color[0][4:])
        if var ==1: 
            step_sub_plot+=1 
            var=0
        else:
            var+=1
            
    sns.despine()
    fig.tight_layout()
    plt.show()
    
def classification_interpretation(y_pred, y_actual): 
    cf_matrix = confusion_matrix(y, y_pred)
    group_names = ['True Neg','False Pos','False Neg','True Pos']
    group_counts = ["{0:0.0f}".format(value) for value in
                    cf_matrix.flatten()]
    group_percentages = ["{0:.2%}".format(value) for value in
                         cf_matrix.flatten()/np.sum(cf_matrix)]
    labels = [f"{v1}\n{v2}\n{v3}" for v1, v2, v3 in
              zip(group_names,group_counts,group_percentages)]
    labels = np.asarray(labels).reshape(2,2)
    sns.heatmap(cf_matrix, annot=labels, fmt='', cmap='hot_r',  linewidths=10)
```

```python
import math
def single_hist(d, col, bn_sz, stat='density'):
    n = math.ceil((d[col].max() - d[col].min())/bn_sz)
    sns.histplot(d,x=col, bins = n, kde=True, color='r', stat=stat)
    sns.despine()
```
